```python
import csv

f = open('C:/Users/lhk65/OneDrive/Desktop/Project/Python_Programming/Part 2/Chapter 12/weather.csv')
data = csv.reader(f)
header = next(data)

max_wind =0.0
for row in data:
    if row[2] =='':
        wind = 0
    else:
        wind = float(row[2])
    if max_wind < wind:
        max_wind = wind

print('지난 10년간 울릉도의 최대 풍속은 ', max_wind, 'm/s')
f.close()
```
